##### **Preprocess the data**

```python
import re
def preprocess(text):
    # remove punctuations
    text = re.sub("[()/.:?,!@#$%\"']", "", text)
    print(text)
    new_string = ""
    for token in text.split():
        # check if no alphabet character
        if re.search("[a-zA-Z]",token):
            # if not in all-caps, make it lower case
            if not token.isupper():
                token = token.lower()
            new_string += (token + " ")
    return new_string
```

```python
test_data_temp = [0]*5
for i in range(len(test_data[:5])):
    test_data_temp[i] = preprocess(test_data[i])
```
